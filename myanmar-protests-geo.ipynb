```python
import pandas as pd
import geopandas as gpd
```

## Load processed incidents data

```python
df = pd.read_csv("data/processed/myanmar-protests-map.csv")

df.info()
df.head()
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df["longitude"], df["latitude"])
)

gdf.info()
gdf.head()
```

## Get bounding box for Mapbox map

```python
print(gdf["geometry"].total_bounds)  # minLng, minLat, maxLng, maxLat
```

## Write to GeoJSON

```python
gdf.to_file("data/processed/myanmar-protests.json", driver="GeoJSON")
```
